```python
weights = [79.8,92.8,70.2,67.5]
heights = [1.78,1.62,1.82,1.70]
students = ["철수","영희", "민지", "민수"]

#bmi = weight / height **2
#18.5 저체중
#122.9 정상
#23.0~ 24.9 과체중
#25.0~ 비만

import numpy as np

np_weights= np.array(weights)
np_heights =  np.array(heights)
bmis = np_weights / np_weights ** 2

plt.bar(students, bmis)
plt.yticks([18.5,22.9,24.9,30], labels=["저체중","정상","과체중","비만"])
```
